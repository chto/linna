In this notebook, we sample a 2d gaussian posterior using LINNA. LINNA isn't designed for low dimension posteiors, so the performance will not be great. However, this notebook illustrates how one can use LINNA to sample posteriors. 

# Create a multivariate gaussian distribution 

```python
ndim = 2
init =  np.random.uniform(size=ndim)
#mean value 
means = np.array([0.1, 1])

ndim = len(init)
#covariance matrix 
cov = np.diag([0.5, 0.2])
#Prior, Theory, and likelihood

priors = []
for i in range(ndim):
    priors.append({
        'param': 'test_{0}'.format(i),
        'dist': 'flat',
        'arg1': -1E4,
        'arg2': 1E4
    })

def theory(x, outdirs):
    x_new = deepcopy(x[1])
    return x_new
```

```python
inv_cov = np.linalg.pinv(cov)
def negloglike(x):
    d = means-x
    return 0.5*d.dot(inv_cov.dot(d))
```

```python
import numdifftools as nd
np.linalg.inv(nd.Hessian(negloglike)(means))
```

# Perform MCMC sampling using Linna

```python
nwalkers = 4 #Number of mcmc walker
pool = None
```

```python
outdir = os.path.abspath(os.getcwd())+"/out/2dgaussian_wide/"
gpunode = None
chain, logprob = ml_sampler(outdir, theory, priors, means, cov, init, pool, nwalkers, gpunode, nepoch=101, nbest=[10000,10000,0,0], chisqcut=1E7)
```
